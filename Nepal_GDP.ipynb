```python
import requests

url = "https://www.worldometers.info/gdp/nepal-gdp/"

html = requests.get(url).content

html
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "html.parser")
```

```python
tb = soup.find('tbody')
```

```python
tb
```

```python
tr = tb.find_all('tr')
```

```python
tr[0]
```

```python
all_data = []
for index in range(0, len(tr)):
    gdp_data = []
    table_row = tr[index]
    table_data = table_row.find_all('td')
    
    for item in table_data:
        gdp_data.append(item.text)
    all_data.append(gdp_data)
```

```python
all_data
```

```python
# taking rows of data as our choice.
```

```python
data = list(map(lambda x : x[0:], all_data))
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df
```

# READ CSV FILE

```python
 df = pd.read_csv('Nepal_GDP')
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['GDP Nomial USD'].str.replace('$', '', regex = True).replace(',',"", regex = True)
```

```python
df['GDP Nomial USD'] = df['GDP Nomial USD'].str.replace('$', '', regex = True).replace(',',"", regex = True).astype(float)
```

```python
df.dtypes
```

```python
df['GDP Real'] = df['GDP Real'].str.replace('$', '', regex = True).replace(',',"", regex = True).astype(float)
```

```python
df.info()
```

```python
df['GDP Change'] = df['GDP Change'].str.replace('%', '', regex = True).astype(float)
```

```python
df.info()
```

```python
df['GDP per Capita'] = df['GDP per Capita'].str.replace('$', '', regex = True).replace(',',"", regex = True).astype(float)
```

```python
df.dtypes
```

```python
df['Population Change'] = df['Population Change'].str.replace('%', '', regex = True).astype(float)
```

```python
df.info()
```

```python
df['Population'] = df['Population'].str.replace(',',"", regex = True).astype(float)
```
